```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
sns.set(style='darkgrid')
```

```python
log_loc = os.path.join('..', 'Data', 'binary_localizer_16_0.28892_1_54_7_12.log')
data = pd.read_csv(log_loc)
plt.plot(data['epoch'], data['loss'])
plt.plot(data['epoch'], data['val_loss'])
plt.legend(['Loss', 'Validation Loss'], fontsize=18)
_=plt.xticks(fontsize=18)
_=plt.yticks(fontsize=18)
# plt.title('Training History')
plt.savefig('training_history.svg', bboxinches='tight')
```

```python
%matplotlib notebook
f, ax = plt.subplots(2,1)
ax[0].plot(data['epoch'], data['average_precision_score'])
ax[0].plot(data['epoch'], data['auc_score'])
ax[0].plot(data['epoch'], data['f1_score'])
ax[0].legend(['AP', 'AUC', 'F1'])
ax[1].plot(data['epoch'], data['loss'], color='purple')
ax[1].plot(data['epoch'], data['val_loss'], color='red')
ax[1].legend(['Loss', 'Validation Loss'])
f.suptitle('Training History')
plt.xlabel('Epoch')
```
